## ggtitle()
Modify axis, legend, and plot labels. Good labels are critical for making your plots accessible to a wider audience. Ensure the axis and legend labels display the full variable name. Use the plot title and subtitle to explain the main findings. It's common to use the caption to provide information about the data source. 
### Usage
ggtitle(label, subtitle = NULL)

```r
# View help on ggtitle()
?ggtitle()
```

```r
# mtcars: y axis: mpg, x axis: disp
# +jitter; +alpha; +linear trend no se
# Title: <y axis> vs <x axis>
# Subtitle: Row count: <nrows>
# Hint: ggtitle(), nrow()
mtcars %>% ggplot(mapping = aes(x =disp, y = mpg)) + 
    geom_jitter(alpha = 0.5) + 
   geom_smooth(method = "lm", se = FALSE) +
   ggtitle("disp vs mpg", 
          subtitle = str_c("Row count:", mtcars %>% nrow()))

# mtcars: y axis: mpg, x axis: hp
# +jitter; +alpha; +linear trend no se
mtcars %>% ggplot(mapping = aes(x =hp, y = mpg)) + 
    geom_jitter(alpha = 0.5) + 
   geom_smooth(method = "lm", se = FALSE) +
   ggtitle("hp vs mpg", 
          subtitle = str_c("Row count:", mtcars %>% nrow()))


# mtcars: y axis: mpg, x axis: wt
# +jitter; +alpha; +linear trend no se
mtcars %>% ggplot(mapping = aes(x =wt, y = mpg)) + 
    geom_jitter(alpha = 0.5) + 
   geom_smooth(method = "lm", se = FALSE) +
   ggtitle("wt vs mpg", 
          subtitle = str_c("Row count:", mtcars %>% nrow()))

```

Let's wrap the ggplot code into a function.

## aes_string()
Define aesthetic mappings programmatically. Aesthetic mappings describe how variables in the data are mapped to visual properties (aesthetics) of geoms. aes uses non-standard evaluation to capture the variable names. aes_ and aes_string require you to explicitly quote the inputs either with "" for aes_string(), or with quote or ~ for aes_(). (aes_q is an alias to aes_). This makes aes_ and aes_string easy to program with. 
### Usage
aes_string(x, y, ...)

```r
# View help on aes_string()
#help(aes_string)

#Examples

# Three ways of generating the same aesthetics
aes(mpg, wt, col = cyl)
aes_(quote(mpg), quote(wt), col = quote(cyl))
aes_(~mpg, ~wt, col = ~cyl)
aes_string("mpg", "wt", col = "cyl")

# You can't easily mimic these calls with aes_string
aes(`$100`, colour = "smooth")
aes_(~ `$100`, colour = "smooth")
# Ok, you can, but it requires a _lot_ of quotes
aes_string("`$100`", colour = '"smooth"')

# Convert strings to names with as.name
var <- "cyl"
aes(col = x)
aes_(col = as.name(var))
```

```r
# Create function gg_scatter
# Pass in data frame as data
# and x_col, y_col as strings for x and y axis columns
# Match functionality of prior code block
# Hint: aes_string()
gg_scatter <- function(data, x_col, y_col) {
    # generate plot stored as plt
    plt <- data %>% ggplot(mapping = aes_string(x = x_col, y = y_col)) + 
       geom_jitter(alpha = 0.5) + 
       geom_smooth(method = "lm", se = FALSE) +
       ggtitle(str_c(y_col, " vs ", x_col), 
              subtitle = str_c("Row count:", data %>% nrow()))
    
    # print plot
    # Hint: print()
    plt %>% print()
}

# Call function for disp, hp, and wt all vs mpg
mtcars%>% gg_scatter("disp", "mpg")
mtcars %>% gg_scatter("hp", "mpg")
mtcars %>% gg_scatter("wt", "mpg")
```

Notice that not only is the ggplot code reused, but it is easier to understand your code with three short function calls instead of three large blocks of ggplot code.

Let's use a For loop to simplify more.

```r
# Define the x axis columns in 
# a character vector x_cols
# for disp, hp, and wt
# Hint: c()
x_cols <- c("disp", "hp", "wt")

# Create a for loop and call gg_scatter
# Use mtcars, x_cols, and "mpg"
for(x_col in x_cols) {
    mtcars %>% gg_scatter(x_col, "mpg")
}
```

Notice the for loop makes it a bit more concise and easier to update the list of columns to plot. 

Let's change the For loop to a walk() functional.

## Walk()
Apply a function to each element of a vector. map() returns the transformed input; walk() calls .f for its side-effect and returns the original input. map() returns a list or a data frame; map_lgl(), map_int(), map_dbl() and map_chr() return vectors of the corresponding type (or die trying); map_df() returns a data frame by row-binding the individual elements. 
### Usage
walk(.x, .f, ...)

```r
# View help on walk()
?walk()
```

```r
# Define the x axis columns in 
# a character vector x_cols
# for disp, hp, and wt
# Hint: c()
x_cols <- c("disp", "hp", "wt")

# Create a functional call to iterate over the objects
# Hint: walk()
 x_cols%>% walk(gg_scatter, data = mtcars, y_col = "mpg")
```

Notice that the For loop is now a single line of code. 

Let's make the x_cols more dynamic and plot all the columns from mtcars.

```r
?names()
```

```r
# Define y axis column as y_col
# y axis column is mpg
y_col <- "mpg"

# Define the x axis columns in 
# a character vector x_cols
# for all columns in mtcars
# Hint: names()
x_cols <- mtcars %>%names()

# Remove y_axis column from x_cols
# Hint: !str_detect()
x_cols <- x_cols[!str_detect(x_cols, y_col)]

# Print x columns
x_cols %>% print()

# Create a functional call to iterate over the objects
# Calling gg_scatter with mtcars and x and y columns
# Hint: walk()
x_cols%>% walk(gg_scatter, data = mtcars, y_col = y_col)
```

Notice, that you can let the data in the data frame drive most of the exploration and creation of charts. 

It might be nice to save the charts as images on disk for later use. Let's add this as a function.

## ggsave()
Save a ggplot (or other grid object) with sensible defaults. ggsave() is a convenient function for saving a plot. It defaults to saving the last plot that you displayed, using the size of the current graphics device. It also guesses the type of graphics device from the extension.
### Usage
ggsave(filename, plot = last_plot(), device = NULL, path = NULL,
  scale = 1, width = NA, height = NA, units = c("in", "cm", "mm"),
  dpi = 300, limitsize = TRUE, ...)

```r
# View help on ggsave()
?ggsave()
```

```r
# Create a function named save_plot
# Parameter: filename
# Function will save the last ggplot
# to the filename
# Hint: ggsave()
save_plot <- function(filename) {
    filename %>% ggsave()
}
```

Notice no output for function definitions. 

Let's test it out. 

```r
# Generate a ggplot
# using gg_scatter with mtcars
# x = disp, y = mpg
mtcars %>% gg_scatter("disp", "mpg")

# call save_plot
# filename: "disp_mpg.png"
save_plot("disp_mpg.png")
```

Notice the warning message that it saved the image. Check the Jupyter Notebook navigation and find the image it just saved and confirm it matches the image that was also output inside the notebook.

Let's move the save_plot function inside the gg_scatter function so it will save the plot to file for every plot automatically.

```r
# Redefine gg_scatter to call save_plot
# Also, move hard coded values to parameters with default values
# Hint: Review prior version of gg_scatter
gg_scatter <- function(data, x_col, y_col, 
                       title_param = "",
                       subtitle_param = "",
                       alpha_param = 0.5, 
                       method_param = 'lm', 
                       se_param = FALSE) {
    # generate plot stored as plt
    plt <- data %>% ggplot(mapping = aes_string(x = x_col, y = y_col)) + 
       geom_jitter(alpha = alpha_param) + 
       geom_smooth(method = method_param, se = se_param) +
       ggtitle(title_param, subtitle = subtitle_param)

    # print plot
    # Hint: print()
    plt %>% print()

    # save plot to file
    # Hint: save_plot()
    str_c(x_col, "_", y_col, ".png") %>% save_plot()
}
```

Notice no visible output of running function definition code. Also notice how the code inside the code block is getting smaller as more of the values are changing to variables. This makes the function more versatile allowing you to override any of the default parameters when necessary.

Let's try out the new function.

```r
# Generate a ggplot
# using gg_scatter with mtcars
# x = hp, y = mpg
mtcars %>% gg_scatter('hp', 'mpg')
```

## Code Recap

```r
# Load libraries
library('stringr')    # stringr
library('tidyverse')  # tidyverse

# Create a function named save_plot
# Parameter: filename
# Function will save the last ggplot
# to the filename
# Hint: ggsave()
save_plot <- function(filename) {
    filename %>% ggsave()
}

# Create function gg_scatter
# Parameters: data = data frame
# Parameters: x_col, y_col as strings for x and y axis columns
# Default params: title, subtitle, alpha, method, se
# Create scatter plot of x and y columms for data
# +jitter, + alpha, +smooth +title
# print plot and save to file
# Hint: Use aes_string(), save_plot()
gg_scatter <- function(data, x_col, y_col, 
                       title_param = str_c(y_col, " vs ", x_col),
                       subtitle_param = str_c("Row count:", data %>% nrow()),
                      alpha_param = 0.5, method_param = "lm", 
                      se_param = FALSE) {
    # generate plot stored as plt
    plt <- data %>% ggplot(mapping = aes_string(x = x_col, y = y_col)) + 
       geom_jitter(alpha = alpha_param) + 
       geom_smooth(method = method_param, se = se_param) +
       ggtitle(title_param, subtitle = subtitle_param)

    # print plot
    # Hint: print()
    plt %>% print()

    # save plot to file
    str_c(x_col, "_", y_col, ".png") %>% save_plot()
}

# Define y axis column as y_col
# y axis column is mpg
y_col <- "mpg"

# Define the x axis columns in 
# a character vector x_cols
# for all columns in mtcars
# Hint: names()
x_cols<- mtcars %>% names()

# Remove y_axis column from x_cols
# Hint: !str_detect()
x_cols <- x_cols[!str_detect(x_cols, y_col)]

# Create a functional call to iterate over the objects
# Calling gg_scatter with mtcars and x and y columns
# Hint: walk()
x_cols%>% walk(gg_scatter, data = mtcars, y_col = y_col)
```
