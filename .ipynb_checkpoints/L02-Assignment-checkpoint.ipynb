```r
# chart 1 placeholder
df_orig %>% ggplot(mapping = aes(x = carat  , y = price, color = color))+
    geom_jitter(alpha = 0.4, height = 0.3)+
    geom_smooth(method = 'lm', se = FALSE)+
    facet_grid(cut~.)+
    labs(title = "Comparing price by Carat")+
    labs(subtitle = "facets are cut, clarity")
```
