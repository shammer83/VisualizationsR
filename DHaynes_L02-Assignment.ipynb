# Chart 1
Optionally double click to edit this text to describe this chart.

```r
?geom_jitter
```

```r
# chart 1 placeholder
df_orig %>%arrange(desc(price))%>% ggplot(mapping = aes(x = carat  , y = price, color = color))+
    geom_jitter(alpha = 0.4, height = 0.3)+
    geom_smooth(method = 'lm', se = FALSE)+
    facet_grid(cut~.)+
    labs(title = "Comparing Price by Carat")+
    labs(subtitle = "Facets: cut")
```

# Chart 2
Optionally double click to edit this text to describe this chart.

```r
# chart 2 placeholder

df_orig %>%arrange(desc(price))%>%ggplot(mapping = aes(x = carat  , y = price, color = clarity))+
    geom_jitter(alpha = 0.4, height = 0.3)+
    geom_smooth(method = 'lm', se = FALSE)+
    facet_grid(cut~.)+
    labs(title = "Assessing Impact of 'Clarity' on Diamond Price")+
    labs(subtitle = "Facet: Cut")
```
